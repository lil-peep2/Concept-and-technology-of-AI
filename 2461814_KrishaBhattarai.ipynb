```python
def classify_hdi_catgory(hdi):
  if hdi < 0.550:
    return "Low"
  elif 0.550 <= hdi <= 0.699:
    return "Medium"
  elif 0.700 <= hdi <= 0.799:
    return "High"
  else:
    return "Very High"
```

#### Which countries show the greatest improvement in HDI from 2020 to 2022?

```python
# keep the values of HDI for the year 2020 and 2022
change_in_hdi = hdi_problem1B_df[hdi_problem1B_df["year"].isin([2020, 2021, 2022])]
```

```python
pivot_hdi = change_in_hdi.pivot(index="country", columns="year", values="hdi")
```

```python
pivot_hdi["hdi_change_2020-2021"] = pivot_hdi[2021] - pivot_hdi[2020]
pivot_hdi["hdi_change_2021-2022"] = pivot_hdi[2022] - pivot_hdi[2021]
```

```python
pivot_hdi["total_hdi_change"] = pivot_hdi["hdi_change_2020-2021"] + pivot_hdi["hdi_change_2021-2022"]
```

```python
greatest_hdi_improvement = pivot_hdi.sort_values(by="total_hdi_change", ascending=False).head(10)
```

```python
print("Countries with greatest improvement in HDI from 2020 to 2022: ")
print(greatest_hdi_improvement)
```

#### Did any countries experience a decline in HDI? Provide possible reasons.

```python
decline_hdi = pivot_hdi[(pivot_hdi["hdi_change_2020-2021"] < 0) | (pivot_hdi["hdi_change_2021-2022"] < 0)]
decline_hdi = decline_hdi.sort_values(by="total_hdi_change")
```

```python
print("The countries with a decline in HDI: ")
print(decline_hdi)
```

#### Discuss how global events (e.g., the COVID-19 pandemic) may have affected HDI trends during this period.

In 2020 and 2021, the COVID-19 pandemic caused a decrease in HDI in the whole world as health services were overloaded, education was interrupted, and millions of individuals lost their sources of income. The wealthy and well-developed countries rebounded faster by 2022 due to the strong healthcare and economic assistance, whereas the poorer and unstable ones experienced delayed recovery. This shows that development should not be based on income alone, but good institutions and policies should be established.

#### Task 3: Outlier Detection

```python
def detect_outliars(df, col):
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  return outliers, lower_bound, upper_bound
```

```python
hdi_outliers, hdi_lower_bound, hdi_upper_bound = detect_outliars(southAsia_hdi_df, "hdi")
```

```python
print("Outliers in HDI: ")
print(hdi_outliers[["country", "year", "hdi"]])
```

```python
gni_outliers, gni_lower_bound, gni_upper_bound = detect_outliars(southAsia_hdi_df, "gross_inc_percap")
```

```python
print("Outliers in GNI per capita: ")
print(gni_outliers[["country", "year", "gross_inc_percap"]])
```

```python
southAsia_hdi_df["outlier"] = (
    (southAsia_hdi_df["hdi"] < hdi_lower_bound) | (southAsia_hdi_df["hdi"] > hdi_upper_bound) |
    (southAsia_hdi_df["gross_inc_percap"] < gni_lower_bound) | (southAsia_hdi_df["gross_inc_percap"] > gni_upper_bound)
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=southAsia_hdi_df, x="gross_inc_percap", y="hdi", hue='outlier', palette={False:"red", True: "Green"},
                markers={False: "o", True: "*"})
plt.title("Scatter Plot of GNI per Capita vs HDI ")
plt.xlabel("Gross National Income (GNI) per Capita")
plt.ylabel("Human Development Index (HDI)")
plt.legend(title="Outlier Status")
plt.grid(True)
caption_text = "Figure: Scatter plot of Gross Income per Capita vs Human Development Index, Outliers detected using 1.5 Ã— IQR rule."
plt.figtext(0.5, 0.01, caption_text, wrap=True, horizontalalignment="center", fontsize=10, color="gray")
plt.show()
```

#### Discuss why the identified countries stand out as outliers.

Maldives (2022) is an outlier since it has a significantly higher GNI per capita in comparison with other South Asian countries.
