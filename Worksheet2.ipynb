### Handling Missing values - Adding Missing Values

```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
import numpy as np
```

```python
iris = load_iris()
```

```python
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])
```

```python
np.random.seed(42)
```

```python
mask = np.random.rand(*iris_df.shape) < 0.1
```

```python
iris_df[mask] = np.nan
```

```python
print("Missing Values in Iris Dataset:")
print(iris_df.isnull().sum())
```

### Handling Missing values - Techniques for Filling Missing Values

```python
iris_df_ffill = iris_df.ffill()
```

```python
iris_df_mean = iris_df.fillna(iris_df.mean())
```

```python
iris_df_median = iris_df.fillna(iris_df.median())
```

```python
iris_df_zero = iris_df.fillna(0)
```

```python
iris_df_expanded = pd.concat([iris_df, iris_df_ffill.add_suffix('_ffill'), iris_df_mean.add_suffix('_mean'),iris_df_median.add_suffix('_median'),iris_df_zero.add_suffix('_zero')], axis=1)
```

```python
print("\nDataset after Filling Missing Values:")
print(iris_df_expanded.head())
```

### DataFrame Reshaping - Pivot and Melt

```python
import pandas as pd
```

```python
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'City': ['Kathmandu', 'Pokhara', 'Kathmandu', 'Pokhara'],
    'Temperature': [15, 18, 16, 19]}
df = pd.DataFrame(data)
```

```python
pivoted_df = df.pivot(index='Date', columns='City', values='Temperature')
print(pivoted_df)
```

```python
melted_df = pd.melt(pivoted_df.reset_index(), id_vars=['Date'], var_name='City', value_name='Temperature')
print(melted_df)
```

### Merging and Joining DataFrames - Merge

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Score': [85, 90, 88]})
```

```python
inner_merged = pd.merge(df1, df2, on='ID', how='inner')
```

```python
print("Inner Join:")
print(inner_merged)
```

```python
left_merged = pd.merge(df1, df2, on='ID', how='left')
```

```python
print("\nLeft Join:")
print(left_merged)
```

```python
outer_merged = pd.merge(df1, df2, on='ID', how='outer')
```

```python
print("\nOuter Join:")
print(outer_merged)
```

Problem 2 - Data Imputations:

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Concept And Technologies of AI/medical_students_dataset.csv")
```

```python
print(df.head())
```

```python
df.info()
```

```python
print("Columns with missing values:/n")
print(df.isnull().sum())
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].median())
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
dup_count = df.duplicated().sum()
print("Total number of duplicate rows:", dup_count)
```

```python
if dup_count > 0:
    df = df.drop_duplicates()
df
```
