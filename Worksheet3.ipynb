### Subsetting by indices

### Creating Subplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
```

```python
plt.figure()
```

```python
plt.subplot(211)
plt.plot(t1, f(t1),'bo', t2, f(t2), 'k')
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:

### What percent of the total fare revenue came from each passenger class?

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Concept And Technologies of AI/Titanic-Dataset.csv")
```

```python
total_fare = df["Fare"].sum()
```

```python
first_class_fare = df[df["Pclass"] == 1]["Fare"].sum()
```

```python
second_class_fare = df[df["Pclass"] == 2]["Fare"].sum()
```

```python
third_class_fare = df[df["Pclass"] == 3]["Fare"].sum()
```

```python
fare_totals = [first_class_fare,second_class_fare,third_class_fare ]
```

```python
proportion_of_fare_revenue = [(fare / total_fare) * 100 for fare in fare_totals]
```

```python
print("Total Fare Revenue: ",total_fare, "%")
```

```python
print("First Class Fare Revenue: ",proportion_of_fare_revenue[0], "%")
print("Second Class Fare Revenue: ",proportion_of_fare_revenue[1], "%")
print("Third Class Fare Revenue: ",proportion_of_fare_revenue[2], "%")
```

### What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Concept And Technologies of AI/Titanic-Dataset.csv")
```

```python
med_age = df["Age"].median()
df["Age"].fillna(med_age, inplace=True)
```

```python
def categorize_passengers(age):
  if age < 18:
    return "child"
  elif age < 65:
    return "adult"
  else:
    return "senior"

df["age_group"] = df['Age'].apply(categorize_passengers)
```

```python
total_passengers = len(df)
print("The total number of passengers is: ",total_passengers)
```

```python
passenger_age_group = df["age_group"].value_counts()
print("The number of passengers in each age group is: ")
print(passenger_age_group)
```

```python
passenger_age_group_percent = (passenger_age_group / total_passengers) * 100
```

```python
print("The percentage of passengers in each age group is: ")
print(passenger_age_group_percent)
```
